```
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from torchvision import datasets
from torchvision import transforms
```

Downloading the CIFAR10 dataset using torchvision datasets and visualising them


```
data_path = '../data-unversioned/p1ch7/'
cifar10_train = datasets.CIFAR10(data_path, train=True, download=True,transform=transforms.ToTensor())
cifar10_test = datasets.CIFAR10(data_path, train=False, download=True,transform=transforms.ToTensor())
classes = cifar10_train.classes

```

```
#randomly generating a number from 1 to len(cifar10_train)
random_index = np.random.randint(len(cifar10_train))
#displaying the image
image , label = cifar10_train[random_index]

plt.imshow(image.permute(1,2,0)) #as imgshow takes HxWxC as its order of parameters
plt.show()
print(classes[label])

```
